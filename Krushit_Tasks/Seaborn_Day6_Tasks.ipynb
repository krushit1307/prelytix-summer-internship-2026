```python
import seaborn as sns
import matplotlib.pyplot as plt

# Task 1.2.1: Load the 'penguins' dataset from Seaborn. Create a scatter plot of bill_length_mm vs bill_depth_mm, colored (hue) by species.
penguins = sns.load_dataset('penguins')
sns.scatterplot(data=penguins, x='bill_length_mm', y='bill_depth_mm', hue='species')
plt.title('Penguin Bill Length vs Depth')
plt.show()
```

```python
# Task 1.2.2: Load the 'flights' dataset. Create a line plot showing the number of passengers over the years. Seaborn will automatically average the months.
flights = sns.load_dataset('flights')
sns.lineplot(data=flights, x='year', y='passengers')
plt.title('Flights Passengers Over Years')
plt.show()
```

```python
# Task 2.2.1: Using the 'titanic' dataset, create a histogram of passenger ages. Add a KDE line and hue by 'survived'.
titanic = sns.load_dataset('titanic')
sns.histplot(data=titanic, x='age', kde=True, hue='survived')
plt.title('Titanic Passengers Age Distribution')
plt.show()
```

```python
# Task 2.2.2: Create a countplot showing the number of passengers in each 'pclass' on the Titanic.
sns.countplot(data=titanic, x='pclass')
plt.title('Passengers by Class')
plt.show()
```

```python
# Task 2.2.3: Create a boxplot showing the 'fare' distribution across different 'pclass' categories.
sns.boxplot(data=titanic, x='pclass', y='fare')
plt.title('Fare Distribution by Class')
plt.show()
```

```python
# Task 3.2.1: Calculate the correlation matrix for the 'tips' dataset. Plot it as a heatmap with annotations (annot=True) and a 'viridis' colormap.
tips = sns.load_dataset('tips')
numeric_tips = tips.select_dtypes(include=['float64', 'int64'])
corr_mat = numeric_tips.corr()

plt.figure(figsize=(8, 6))
sns.heatmap(corr_mat, annot=True, cmap='viridis')
plt.title('Tips Correlation Heatmap')
plt.show()
```

```python
# Task 3.2.2: Create a pairplot for the 'iris' dataset, coloring the points by 'species'. Which species is most easily separable?
iris = sns.load_dataset('iris')
sns.pairplot(iris, hue='species')
plt.show()
# setosa is typically the most easily separable species in the iris dataset.
```
