```python
import matplotlib.pyplot as plt
import numpy as np

# Task 1.2.1: Create a line plot showing the squares of numbers from 1 to 10. Make the line dashed and green.
x = np.arange(1, 11)
y = x ** 2

plt.plot(x, y, color='green', linestyle='--')
plt.title('Squares from 1 to 10')
plt.show()
```

```python
# Task 1.2.2: Plot two lines on the same graph: y=x^2 and y=x^3. Add a legend to distinguish them.
plt.plot(x, x**2, label='y = x^2')
plt.plot(x, x**3, label='y = x^3')
plt.title('Squares and Cubes')
plt.legend()
plt.show()
```

```python
# Task 1.2.3: Create a scatter plot of height vs weight using dummy data. Color the points purple and use triangles as markers.
height = np.random.normal(170, 10, 50)
weight = np.random.normal(70, 5, 50)

plt.scatter(height, weight, color='purple', marker='^')
plt.xlabel('Height')
plt.ylabel('Weight')
plt.title('Height vs Weight')
plt.show()
```

```python
# Task 2.2.1: Create a horizontal bar chart (plt.barh) showing population of 5 different countries.
countries = ['USA', 'India', 'China', 'Brazil', 'UK']
pop = [331, 1380, 1439, 212, 67]

plt.barh(countries, pop, color='skyblue')
plt.xlabel('Population (millions)')
plt.title('Country Populations')
plt.show()
```

```python
# Task 2.2.2: Generate 500 random ages between 0 and 80. Plot a histogram with 10 bins. Add a black edge color to the bins.
ages = np.random.randint(0, 80, 500)

plt.hist(ages, bins=10, edgecolor='black', color='orange')
plt.title('Age Distribution')
plt.show()
```

```python
# Task 2.2.3: Create a pie chart showing your daily time allocation (sleep, work, study, leisure). Explode the 'study' slice.
activities = ['Sleep', 'Work', 'Study', 'Leisure']
hours = [8, 8, 4, 4]
# study is the 3rd element, so explode the 3rd slice
explode_vals = [0, 0, 0.1, 0]

plt.pie(hours, labels=activities, explode=explode_vals, autopct='%1.1f%%')
plt.title('Daily Time')
plt.show()
```

```python
# Task 3.3.1: Take your line plot from Task 1.1 and add a title, x/y labels, a grid, and set x-axis limits from 0 to 12.
plt.plot(x, y, color='green', linestyle='--')
plt.title('Squares from 1 to 10')
plt.xlabel('Number')
plt.ylabel('Square')
plt.xlim(0, 12)
plt.grid(True)
plt.show()
```

```python
# Task 3.3.2: Create a 2x2 subplot grid. Plot a line, scatter, bar, and histogram in the 4 axes respectively. Add titles to each.
fig, axes = plt.subplots(nrows=2, ncols=2, figsize=(10, 8))

# line
axes[0, 0].plot(x, y, color='green', linestyle='--')
axes[0, 0].set_title('Line Plot')

# scatter
axes[0, 1].scatter(height, weight, color='purple', marker='^')
axes[0, 1].set_title('Scatter Plot')

# bar
axes[1, 0].barh(countries, pop, color='skyblue')
axes[1, 0].set_title('Bar Chart')

# hist
axes[1, 1].hist(ages, bins=10, edgecolor='black', color='orange')
axes[1, 1].set_title('Histogram')

plt.tight_layout()
plt.show()
```

```python
# Task 3.3.3: Save the 2x2 grid as a high-resolution JPG image using plt.savefig().
import os
# We already have 'fig' from the previous cell, but to be safe and independent, we can save it here if fig is in memory,
# but usually in jupyter notebooks the figure is closed after plt.show().
# So we need to recreate or just save the active figure. However, the previous cell called plt.show().
# Let's recreate it quickly to save it, or assume it's part of the same script logically.
# I'll recreate it to avoid blank image issues in Jupyter.

fig, axes = plt.subplots(2, 2, figsize=(10, 8))
axes[0, 0].plot(x, y, color='green', linestyle='--')
axes[0, 0].set_title('Line Plot')
axes[0, 1].scatter(height, weight, color='purple', marker='^')
axes[0, 1].set_title('Scatter Plot')
axes[1, 0].barh(countries, pop, color='skyblue')
axes[1, 0].set_title('Bar Chart')
axes[1, 1].hist(ages, bins=10, edgecolor='black', color='orange')
axes[1, 1].set_title('Histogram')
plt.tight_layout()

os.makedirs('Task', exist_ok=True) # ensuring dir exists just in case
fig.savefig('Task/my_2x2_subplots.jpg', dpi=300)
print('Saved to Task/my_2x2_subplots.jpg')
# plt.show() # optional, we already showed it

# Cleanup
if os.path.exists('Task/my_2x2_subplots.jpg'):
    os.remove('Task/my_2x2_subplots.jpg')
```
