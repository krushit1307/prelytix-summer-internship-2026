```python
import pandas as pd
import numpy as np
import seaborn as sns

# Task 2.2.1: Create a Series of 7 temperatures (float) with days of the week as index. Find the hottest and coldest day using .idxmax() and .idxmin().
temps = pd.Series([22.5, 25.1, 28.0, 24.5, 20.1, 19.5, 23.0], 
                  index=['Mon', 'Tue', 'Wed', 'Thu', 'Fri', 'Sat', 'Sun'])
hottest = temps.idxmax()
coldest = temps.idxmin()
print(f"Hottest: {hottest} ({temps[hottest]})")
print(f"Coldest: {coldest} ({temps[coldest]})")
```

```python
# Task 2.2.2: Create two Series with overlapping indices. Add them — explain where NaN appears and how to fill it using .add(fill_value=0).
s1 = pd.Series([10, 20, 30], index=['A', 'B', 'C'])
s2 = pd.Series([5, 15, 25], index=['B', 'C', 'D'])
# NaN appears at A and D because they don't exist in both series
print("Direct addition:\n", s1 + s2)
print("Filled addition:\n", s1.add(s2, fill_value=0))
```

```python
# Task 2.2.3: From the Series [45,82,67,55,91,38,74], filter all values above the mean. Print both the mask and the filtered values.
s = pd.Series([45, 82, 67, 55, 91, 38, 74])
mean_val = s.mean()
mask = s > mean_val
print("Mean:", mean_val)
print("Mask:\n", mask)
print("Filtered:\n", s[mask])
```

```python
# Task 3.3.2: Load the Titanic dataset (seaborn). Print shape, dtypes, and count of missing values per column.
titanic = sns.load_dataset('titanic')
print("Shape:", titanic.shape)
print("Dtypes:\n", titanic.dtypes)
print("Missing Values:\n", titanic.isnull().sum())
```

```python
# Task 3.3.3: Use describe() to find: which numeric column has the highest standard deviation? Which has the most skewed distribution?
desc = titanic.describe()
std_row = desc.loc['std']
highest_std = std_row.idxmax()
print("Highest std dev is in column:", highest_std)
# For skewness, we can compare mean and median (50%)
mean_median_diff = abs(desc.loc['mean'] - desc.loc['50%'])
most_skewed = mean_median_diff.idxmax()
print("Most skewed column (based on mean-median diff) is:", most_skewed)
```

```python
# Task 3.3.4: Write a function df_summary(df) that prints: shape, column names, dtypes, null count per column, and first 3 rows.
def df_summary(df):
    print("Shape:", df.shape)
    print("Columns:", df.columns.tolist())
    print("DTypes:\n", df.dtypes)
    print("Null Counts:\n", df.isnull().sum())
    print("First 3 rows:\n", df.head(3))

df_summary(titanic)
```

```python
# Task 5.2.1: On the Titanic dataset: (a) find % missing per column, (b) drop columns with >50% missing, (c) fill Age with median grouped by Pclass.
pct_miss = titanic.isnull().sum() / len(titanic) * 100
print("% missing:\n", pct_miss)

cols_to_drop = pct_miss[pct_miss > 50].index
df_clean = titanic.drop(columns=cols_to_drop)

# fill age with median grouped by pclass
df_clean['age'] = df_clean.groupby('pclass')['age'].transform(lambda x: x.fillna(x.median()))
print("Missing age after fill:", df_clean['age'].isnull().sum())
```

```python
# Task 5.2.2: Create a DataFrame with intentional NaN values. Compare dropna(), fillna(0), fillna(mean), and interpolate() — show the result of each.
df_nan = pd.DataFrame({'A': [1, np.nan, 3, 4], 'B': [np.nan, 2, np.nan, 5]})
print("Original:\n", df_nan)
print("dropna:\n", df_nan.dropna())
print("fillna(0):\n", df_nan.fillna(0))
print("fillna(mean):\n", df_nan.fillna(df_nan.mean()))
print("interpolate:\n", df_nan.interpolate())
```

```python
# Task 5.2.3: Use df.fillna(df.mean()) to fill all numeric nulls at once. Then use df.fillna(df.mode().iloc[0]) for non-numeric nulls.
num_cols = titanic.select_dtypes(include='number').columns
non_num_cols = titanic.select_dtypes(exclude='number').columns

df_filled = titanic.copy()
df_filled[num_cols] = df_filled[num_cols].fillna(df_filled[num_cols].mean())
df_filled[non_num_cols] = df_filled[non_num_cols].fillna(df_filled[non_num_cols].mode().iloc[0])

print("Nulls after fill:\n", df_filled.isnull().sum())
```

```python
# Task 6.1.1: On Titanic: add FamilySize, IsAlone (1/0), and AgeGroup ('Child'/<18, 'Adult'/18-60, 'Senior'>60).
df_ops = titanic.copy()
df_ops['FamilySize'] = df_ops['sibsp'] + df_ops['parch'] + 1
df_ops['IsAlone'] = (df_ops['FamilySize'] == 1).astype(int)

bins = [0, 18, 60, 120]
labels = ['Child', 'Adult', 'Senior']
df_ops['AgeGroup'] = pd.cut(df_ops['age'], bins=bins, labels=labels)
print(df_ops[['sibsp', 'parch', 'FamilySize', 'IsAlone', 'age', 'AgeGroup']].head())
```

```python
# Task 6.1.2: Rename all Titanic columns to TitleCase. Then reset them back to lowercase using a single rename call.
df_ops = df_ops.rename(columns=lambda c: c.title())
print("Title Case:", df_ops.columns.tolist())

df_ops = df_ops.rename(columns=lambda c: c.lower())
print("lower case:", df_ops.columns.tolist())
```

```python
# Task 6.1.3: Convert the 'fare' column from float64 to int. What information is lost? How would you round instead of truncate?
# Converting to int truncates the decimal part, losing precision.
df_ops['fare_trunc'] = df_ops['fare'].astype(int)
df_ops['fare_round'] = df_ops['fare'].round().astype(int)
print(df_ops[['fare', 'fare_trunc', 'fare_round']].head())
```

```python
# Task 6.1.4: Drop all columns that have any missing values. Print shape before and after.
print("Shape before:", df_ops.shape)
df_no_nulls = df_ops.dropna(axis=1)
print("Shape after:", df_no_nulls.shape)
```
